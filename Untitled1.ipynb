## Part 1: Printing Right Triangles with Hashtags
Accept the edge length of the base of the right triangle. Print the triangle with hashtags in two different ways. If 5 is given by the user as an input, the program will print hash tags (#) starting with one hashtag up to five to create the triangle. Then, it will print a triangle starting with 5 hashtags down to one.

```python
while True:
    edgelength = int(input("Please enter the edge length of the base of right triangle: "))
    #ask the user for the length
    if edgelength <= 0:
        print("All done!")
        break
     #If user enter 0 or a negative number, the loops ends and the following message is printed: "All done!"
    for i in range (1, edgelength +1):   
         print ("#" * i)
     # Print triangle growing from 1 to the edgelength    
    print() 
    # Print the blank line as guided
    for i in range( edgelength,0, -1):
         print ("#" *i)
     # Print triangle growing from the edgelength to 1
```

## Part 2: Calculate basic statistics for these values 9, 11, 22, 34, 17, 22, 34, 22 and 40.  Print a label and the values for the following: mean, median and mode, sum, minimum, and maximum. Your output should be easy to read and well formatted.



```python
data = [9, 11, 22, 34, 17, 22, 34, 22, 40]
mean = sum(data) / len(data)
#Dividing the sum by the length gives the average
median = sorted(data) [len(data) //2]
## Find the median by first sorting the list.
# Then use integer division (len(data) // 2) to get the index of the middle element.
# Finally select that middle element from the sorted list.
mode = max(data, key= data.count)
#finds the value that appears the most (the mode).
sum_val = sum(data)
#find the sum
min_val = min(data)
#find the min value
max_val = max(data)
#find the max value
print("Basic Statistics for the dataset:")
print("Mean: ", mean)
print("Median: ", median)
print("Mode: ", mode)
print("Sum: ", sum_val)
print("Min: ", min_val)
print("Max: ", max_val)
```
